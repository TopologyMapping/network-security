```python
import pandas as pd
import pickle
```

```python
FILE_NAME_IN = 'orgsResult_v2.pkl'
FILE_NAME_OUT ='organizations.csv'
```

```python
with open(FILE_NAME_IN, 'rb') as file:
    data = pickle.load(file)

pickle_data = []
for uuid, (ip_str, port, org, org_name, industry_info) in data.items():
    pickle_data.append([uuid, ip_str, port, org, org_name, industry_info])
pickle_df = pd.DataFrame(pickle_data, columns=['uuid', 'ip_str', 'port', 'org', 'org_name', 'industry_info'])

if 'industry_info' in pickle_df.columns:
    industry_info_expanded = pd.json_normalize(pickle_df['industry_info'])

    pickle_df = pd.concat([pickle_df.drop(columns=['industry_info']), industry_info_expanded], axis=1)

pickle_df.head()
```

```python
pickle_df.to_csv(FILE_NAME_OUT, index=False)
```
