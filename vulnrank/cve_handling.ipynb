```python
import pandas as pd
import pickle
```

```python
FILE_NAME_IN = 'cvesResult.pkl'
FILE_NAME_OUT = 'vulnerabilities.csv'
```

```python
with open(FILE_NAME_IN, 'rb') as file:
    data = pickle.load(file)

df = pd.DataFrame.from_dict(data).T.reset_index()

df.columns = ['cve_id', 'Description', 'Other Columns']

print(df.head())
```

```python
vulnerability_df = df['Other Columns'].apply(pd.Series)

df = pd.concat([df.drop(columns=['Other Columns']), vulnerability_df], axis=1)

df.columns = ['cve_id', 'Description', 'Denial of Service', 'Remote Code Execution', 'Information Disclosure', 'Privilege Escalation', 'Cross Site Scripting', 'Buffer Overflow', 'Cross Site Request Forgery', 'SQL Injection']

df.head()
```

```python
df.to_csv(FILE_NAME_OUT, index=False)
```
